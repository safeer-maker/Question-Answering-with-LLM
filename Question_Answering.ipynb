```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer

# Define your question and context
question = "How many parameters does BERT-large have?"
context = "BERT-large is really big... it has 24-layers and an embedding size of 1,024, for a total of 340M parameters! Altogether it is 1.34GB, so expect it to take a couple minutes to download to your Colab instance."

# Initialize the model and tokenizer
checkpoint = "bert-large-uncased-whole-word-masking-finetuned-squad"
model = AutoModelForQuestionAnswering.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

# Tokenize the inputs
inputs = tokenizer(question, context, return_tensors="pt", truncation=True, padding=True)

# Perform question answering
start_scores, end_scores = model(**inputs)

print("----- ", start_scores, end_scores)

# Find the answer span with the highest scores
start_index = start_scores.argmax(dim=1).item()
end_index = end_scores.argmax(dim=1).item()

# Get the answer text from the original context
answer = context[start_index:end_index+1]

print("Answer:", answer)
```

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer

# Define your question and context
question = "How does the BART model perform on question answering?"
context = "The BART model is a powerful sequence-to-sequence model that can be fine-tuned for question answering tasks. It has achieved state-of-the-art performance on various question answering datasets."

# Initialize the model and tokenizer
checkpoint = "phiyodr/bart-large-finetuned-squad2"
model = AutoModelForQuestionAnswering.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

# Tokenize the inputs
inputs = tokenizer(question, context, return_tensors="pt", truncation=True, padding=True)

# Perform question answering
start_scores, end_scores, _ = model(**inputs)

# Find the answer span with the highest scores
start_index = start_scores.argmax(dim=1).item()
end_index = end_scores.argmax(dim=1).item()

# Get the answer text from the original context
answer = context[start_index:end_index+1]

print("Answer:", answer)
```
